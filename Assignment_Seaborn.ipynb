# Ans1-

Scatter Plot: Display relationship between two numerical variables.

Bar Plot: Compare categories across a single or multiple variables.

Histogram:Show distribution of a single variable.

Box Plot: Visualize distribution and identify outliers in a variable.

Heatmap:Display matrix-like data, useful for correlation matrices or 2D data representation.

# Ans2-

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.show()

sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.show()
```

# Ans4-

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from seaborn
diamonds = sns.load_dataset('diamonds')

# Plot histogram with 'cut' as hue
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices with Cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')

# Plot pair plot with 'species' as hue
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.title('Pair Plot of Iris Dataset with Species Hue')
plt.show()
```

# Ans6-

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from seaborn
flights = sns.load_dataset('flights')

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Passenger Count Heatmap for Flights Dataset')
plt.show()
```
